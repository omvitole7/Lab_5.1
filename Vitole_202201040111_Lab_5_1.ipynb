### 📌 **Paper Summary:**

**Title:** Learning Transferable Visual Models From Natural Language Supervision  
**Authors:** Alec Radford et al.  
**Date:** March 2020  
**arXiv ID:** 2003.05672v1

---

### ✅ **Objective:**
The paper introduces **CLIP (Contrastive Language–Image Pretraining)**, a method for learning a **vision model** using **natural language supervision** instead of traditional labeled datasets. The goal is to create **generalizable vision models** that can perform well across many tasks without task-specific training.

---

### 🔧 **Methodology:**

- CLIP learns to **match images and text** using a **contrastive learning objective**.
- It is trained on a **large dataset of 400 million (image, text) pairs** collected from the internet.
- The model comprises:
  - A **vision encoder** (ResNet or Vision Transformer),
  - A **text encoder** (Transformer),
  - Both are trained to **project image and text embeddings into a shared space**, where matching pairs are closer together.

---

### 💡 **Key Concepts:**

- **Zero-Shot Learning:** Once trained, CLIP can perform tasks like classification by comparing image embeddings to a set of class label embeddings — no task-specific fine-tuning is needed.
- **Natural Language Supervision:** Enables the model to generalize across tasks, as text descriptions are more flexible than fixed labels.

---

### 📊 **Experiments & Results:**

- CLIP is evaluated on **over 30 datasets** covering various domains.
- **Outperforms** previous zero-shot models and sometimes rivals fully supervised models.
- Shows **strong performance on fine-grained and broad classification tasks**.

---

### 🧠 **Findings:**

- CLIP demonstrates the **scalability and versatility** of training vision models using internet-scale text supervision.
- It **reduces the need for manual labeling**.
- Models trained this way can be **used out-of-the-box** on new tasks with competitive performance.

---

### 🔚 **Conclusion:**

CLIP is a **powerful and flexible vision-language model** that aligns images and text into a shared embedding space, enabling **general-purpose zero-shot image classification** and demonstrating the potential of **natural language supervision** for computer vision.


# **LSTM Implementation**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Load the dataset
file_path = "/content/timesData.csv"
data = pd.read_csv(file_path)

# Set year as index
data.set_index("year", inplace=True)

# Selecting a numerical column for forecasting (change this as per requirement)
target_column = "teaching"  # Replace with actual column
data = data[[target_column]].dropna()  # Drop NaN values

# Visualizing the selected data
plt.figure(figsize=(10, 5))
plt.plot(data, label="Actual Data")
plt.title(f"Time Series Data ({target_column})")
plt.legend()
plt.show()

# Feature Scaling
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)

# Creating sequences for LSTM
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i : i + seq_length])
        y.append(data[i + seq_length])
    return np.array(X), np.array(y)

seq_length = 10  # Adjust as needed
X, y = create_sequences(scaled_data, seq_length)

# Splitting into training and testing sets
train_size = int(len(X) * 0.8)
X_train, y_train = X[:train_size], y[:train_size]
X_test, y_test = X[train_size:], y[train_size:]

# Reshaping data for LSTM (samples, time steps, features)
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

# LSTM Model Definition
model = Sequential([
    LSTM(50, activation="relu", return_sequences=True, input_shape=(seq_length, 1)),
    Dropout(0.2),
    LSTM(50, activation="relu"),
    Dropout(0.2),
    Dense(1)
])

model.compile(optimizer="adam", loss="mse")
model.summary()

# Training the Model
history = model.fit(X_train, y_train, epochs=50, batch_size=16, validation_data=(X_test, y_test), verbose=1)

# Plot Loss
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.legend()
plt.show()

# Predictions
y_pred = model.predict(X_test)
y_pred = scaler.inverse_transform(y_pred)
y_test = scaler.inverse_transform(y_test.reshape(-1, 1))

# Evaluation
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")

# Plot Predictions vs Actual Data
plt.figure(figsize=(10, 5))
plt.plot(y_test, label="Actual")
plt.plot(y_pred, label="Predicted")
plt.title("LSTM Forecasting Results")
plt.legend()
plt.show()
```
